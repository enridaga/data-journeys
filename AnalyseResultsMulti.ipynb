```python
def Table(dataframe, dataset, test_regime):
    # VP, TestRegime 1
    df = dataframe[dataframe['Dataset'] == dataset].drop(columns=['Notebooks','Dataset'])
    TR = df.loc[df['TestRegime']==test_regime].drop(columns=['TestRegime'])
    return TR.groupby(["EmbMethod",'InputSize','Method']).agg(['mean','std'])
```

## Test regime = 1

1. Train and test entities randomly split
2. Train and test do not share notebooks
3. Train and test do not have nodes sharing the same direct arc

```python
Table(data, 'MultiClassification.csv', 1 )
```
